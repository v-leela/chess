```python
import pandas as pd
import glob
import os
```

```python
project_dir = os.path.dirname(os.getcwd())
```

```python
all_files = glob.glob(os.path.join(project_dir, "data/raw_data/games_20*.csv"))
all_files = sorted(all_files)
```

Combining all games

```python
all_month_df = []
for file in all_files:
    df_month = pd.read_csv(file)
    all_month_df.append(df_month)

games = pd.concat(all_month_df, ignore_index=True)
```

Handling duplicate rows and nan

```python

games = games.drop_duplicates()
games["ECO"] = games["ECO"].fillna("Unknown")
games["Opening"] = games["Opening"].fillna("Unknown")
games["Moves"] = games["Moves"].fillna("Unknown")
games.dropna(inplace=True, ignore_index=True)
print(games.isna().sum())
print(len(games))
```

<h>Removing abnormally elo rated rows</h>
    <p>The elo > 3000 seems like the elo rating of chess engine</p>


```python
print(games.loc[games["BlackElo"].idxmax()])
print((games["WhiteElo"] >= 3000).sum())
print((games["BlackElo"] >= 3000).sum())
games=games[(games["BlackElo"] < 3000) & (games["WhiteElo"] < 3000)]
print(len(games))
print(games.loc[games["BlackElo"].idxmax()])
```

Converting the Date column to Datetime

```python
games["Date"] = pd.to_datetime(games["Date"], format="%Y.%m.%d", errors="coerce")
print(games.isna().sum())
games=games[games["Date"].notna()]
games=games.sort_values("Date").reset_index(drop=True)
```

Clening the Results Column

```python
games["Result"].value_counts()
```

```python
games["Result"] = games["Result"].replace("1/2 1/2", "1/2-1/2").replace("1-O", "1-0").replace("00-1", "0-1")
valid_results=["1-0", "0-1", "1/2-1/2"]
games = games[games["Result"].isin(valid_results)]
games["Result"].value_counts()
```

```python
games["Result"] = games["Result"].map({"1-0": 1, "1/2-1/2": 0.5, "0-1": 0})
games["Result"].value_counts()
```

Final check

```python
print(games.shape)

print(games.isna().sum())

print(games[games.duplicated()])
games = games.drop_duplicates()
print(games.duplicated().sum())

print(games["Result"].value_counts())

print(games.describe())
```

Converting the Dataframe to .csv

```python
games.info(memory_usage="deep")
games.drop(columns=["Moves"],inplace=True)
games.info()
```

```python
games.to_csv(os.path.join(project_dir, "data/cleaned_chess.csv"), index=False)
```
